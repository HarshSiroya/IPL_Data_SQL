```python
create column role_id - when Is_Keeper = 0 and Is_Captain = 0 assign it value of 3, when Is_Keeper = 1 and Is_Captain = 0 assign it value 2, when Is_Keeper = 0 and Is_Captain = 1, assign it value 1 and when Is_Captain =1 and Is_Keeper = 1 assign it value 4
```

```python
conditions = [
    (player_matches['Is_Captain'] == 0) & (player_matches['Is_Keeper'] == 0),
    (player_matches['Is_Captain'] == 0) & (player_matches['Is_Keeper'] == 1),
    (player_matches['Is_Captain'] == 1) & (player_matches['Is_Keeper'] == 0),
    (player_matches['Is_Captain'] == 1) & (player_matches['Is_Keeper'] == 1)
]
values = [3, 2, 1, 4]

player_matches['role_id'] = np.select(conditions, values, default=np.nan)
```

```python
player_matches
```

```python
player_matches['role_id'] = player_matches['role_id'].astype('Int64')
```

```python
player_matches.drop("Is_Keeper", axis=1, inplace=True)
```

```python
player_matches.drop("Is_Captain", axis=1, inplace=True)
```

```python
player_matches
```

```python
player_matches = player_matches[['Match_Id', 'Player_Id', 'role_id', 'Team_Id']]
```

```python
player_matches
```

```python
player_matches = player_matches.rename(columns={
    'Match_Id': 'match_id',
    'Player_Id': 'player_id',
    'Team_Id': 'team_id'
})
```
